```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Baca data dari file
data_nasabah = pd.read_csv('data_nasabah.csv')

# Membuat subplots
fig, axes = plt.subplots(4, 3, figsize=(18, 20))

# bins untuk mengkategorikan usia
bins = [20, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']
data_nasabah['Usia_Kategori'] = pd.cut(data_nasabah['Usia_Nasabah'], bins=bins, labels=labels, right=False)

# Jumlah nasabah berdasarkan kategori usia
jumlah_nasabah_per_kategori = data_nasabah['Usia_Kategori'].value_counts().sort_index()
axes[0, 0].bar(jumlah_nasabah_per_kategori.index, jumlah_nasabah_per_kategori.values, color='lightgreen')
axes[0, 0].set_title('Jumlah Nasabah Berdasarkan Kategori Usia')
axes[0, 0].set_xlabel('Kategori Usia')
axes[0, 0].set_ylabel('Jumlah Nasabah')
axes[0, 0].tick_params(axis='x', rotation=45)

# Persentase Jenis Kelamin Nasabah
data_nasabah['Jenis_Kelamin'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], ax=axes[0, 1])
axes[0, 1].set_title('Persentase Jenis Kelamin Nasabah')
axes[0, 1].set_ylabel('')

# Jumlah Tanggungan Nasabah
jumlah_tanggungan = data_nasabah['Jumlah_Tanggungan'].value_counts()
axes[0, 2].plot(jumlah_tanggungan.index, jumlah_tanggungan.values, marker='o', linestyle='-')
axes[0, 2].set_title('Jumlah Tanggungan Nasabah')
axes[0, 2].set_xlabel('Jumlah Tanggungan')
axes[0, 2].set_ylabel('Jumlah Nasabah')
axes[0, 2].set_xticks(jumlah_tanggungan.index)

# Tingkat Pendidikan
jumlah_pendidikan = data_nasabah['Tingkat_Pendidikan'].value_counts()
jumlah_pendidikan.plot(kind='bar', color='skyblue', ax=axes[1, 0])
axes[1, 0].set_title('Jumlah Nasabah Berdasarkan Tingkat Pendidikan')
axes[1, 0].set_xlabel('Tingkat Pendidikan')
axes[1, 0].set_ylabel('Jumlah Nasabah')
axes[1, 0].tick_params(axis='x', rotation=45)

# Status Pernikahan
jumlah_status_pernikahan = data_nasabah['Status_Pernikahan'].value_counts()
jumlah_status_pernikahan.plot(kind='bar', color='skyblue', ax=axes[1, 1])
axes[1, 1].set_title('Distribusi Status Pernikahan Nasabah')
axes[1, 1].set_xlabel('Status Pernikahan')
axes[1, 1].set_ylabel('Jumlah Nasabah')
axes[1, 1].tick_params(axis='x', rotation=45)

# Kategori Kartu
jumlah_kartu = data_nasabah['Kategori_Kartu'].value_counts()
jumlah_kartu.plot(kind='bar', color='skyblue', ax=axes[1, 2])
axes[1, 2].set_title('Jumlah Nasabah Berdasarkan Kategori Kartu')
axes[1, 2].set_xlabel('Kategori Kartu')
axes[1, 2].set_ylabel('Jumlah Nasabah')
axes[1, 2].tick_params(axis='x', rotation=45)

# Lama Menjadi Nasabah
axes[2, 0].hist(data_nasabah['Lama_Menjadi_Nasabah'], bins=20, color='skyblue', edgecolor='black')
axes[2, 0].set_title('Distribusi Lama Menjadi Nasabah')
axes[2, 0].set_xlabel('Lama Menjadi Nasabah (Tahun)')
axes[2, 0].set_ylabel('Jumlah Nasabah')

# Jumlah Layanan
axes[2, 1].hist(data_nasabah['Jumlah_Layanan'], bins=20, color='orange', edgecolor='black')
axes[2, 1].set_title('Distribusi Jumlah Layanan yang Diterima oleh Nasabah')
axes[2, 1].set_xlabel('Jumlah Layanan')
axes[2, 1].set_ylabel('Jumlah Nasabah')

# Rata-Rata Limit Kredit Berdasarkan Kategori Kartu
rata_rata_limit = data_nasabah.groupby('Kategori_Kartu')['Limit_Kredit'].mean().sort_values()
rata_rata_limit.plot(kind='bar', color='skyblue', ax=axes[2, 2])
axes[2, 2].set_title('Rata-Rata Limit Kredit Berdasarkan Kategori Kartu')
axes[2, 2].set_xlabel('Kategori Kartu')
axes[2, 2].set_ylabel('Rata-Rata Limit Kredit')
axes[2, 2].tick_params(axis='x', rotation=45)

# Saldo Revolting Total
axes[3, 0].hist(data_nasabah['Saldo_Revolting_Total'], bins=20, color='skyblue', edgecolor='black')
axes[3, 0].set_title('Distribusi Saldo Revolting Total Nasabah')
axes[3, 0].set_xlabel('Saldo Revolting Total')
axes[3, 0].set_ylabel('Jumlah Nasabah')

# Total Transaksi
axes[3, 1].hist(data_nasabah['Total_Transaksi'], bins=20, color='skyblue', edgecolor='black')
axes[3, 1].set_title('Distribusi Total Transaksi Nasabah')
axes[3, 1].set_xlabel('Total Transaksi')
axes[3, 1].set_ylabel('Jumlah Nasabah')

# Jumlah Melakukan Transaksi
axes[3, 2].hist(data_nasabah['Jumlah_Melakukan_Transaksi'], bins=50, color='skyblue', edgecolor='black')
axes[3, 2].set_title('Distribusi Jumlah Melakukan Transaksi Nasabah')
axes[3, 2].set_xlabel('Jumlah Melakukan Transaksi')
axes[3, 2].set_ylabel('Jumlah Nasabah')

# Rasio Penggunaan Rata-rata
sns.kdeplot(data_nasabah['Rasio_Penggunaan_Rata-rata'], color='blue', shade=True, ax=axes[3, 2])
axes[3, 2].set_title('Distribusi Rasio Penggunaan Rata-rata Kartu Kredit')
axes[3, 2].set_xlabel('Rasio Penggunaan Rata-rata')
axes[3, 2].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

Berdasarkan analisis data, mayoritas nasabah dalam dataset ini memiliki rentang usia antara 40 hingga 49 tahun. Meskipun mayoritas nasabah adalah perempuan, perbedaannya sangat tipis dengan jumlah nasabah laki-laki. Secara umum, nasabah memiliki jumlah tanggungan sebanyak 4 orang, dan mayoritas dari mereka memiliki tingkat pendidikan sarjana. Mayoritas nasabah juga sudah menikah, dengan jumlah yang signifikan dibandingkan dengan nasabah yang belum menikah atau cerai. Kartu biru adalah kategori kartu yang paling banyak dimiliki oleh nasabah, sementara kartu platinum jumlahnya paling sedikit. Limit kredit tertinggi umumnya diperoleh oleh pengguna kartu platinum, melebihi 30000. Sedangkan jumlah nasabah dengan saldo revolting di atas 1500 terbilang sedikit. Lebih dari 250 nasabah melakukan transaksi dengan total nominal di atas 5000, sementara sebagian besar nasabah melakukan transaksi antara 60 hingga 80 kali. Rasio penggunaan kartu kredit rata-rata masih perlu dipahami lebih lanjut. Dengan demikian, dapat disimpulkan bahwa nasabah dalam dataset ini umumnya adalah individu dengan usia menengah, sudah menikah, memiliki pendidikan tinggi, dan mayoritas menggunakan kartu biru dengan sebagian kecil menggunakan kartu platinum, sementara kebiasaan transaksinya bervariasi.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Kolom untuk boxplot
col_box = ['Usia_Nasabah', 'Jumlah_Tanggungan', 'Pendapatan',
           'Lama_Menjadi_Nasabah', 'Jumlah_Layanan',
           'Limit_Kredit', 'Saldo_Revolting_Total', 'Total_Transaksi',
           'Jumlah_Melakukan_Transaksi', 'Rasio_Penggunaan_Rata-rata']

# Palet warna untuk setiap subplot
color_palette = sns.color_palette("Set1", len(col_box))

# Membuat subplot untuk boxplot
plt.figure(figsize=(28, 24))
for i, column in enumerate(col_box, 1):
    plt.subplot((len(col_box) // 3) + 1, 3, i)
    sns.histplot(data=dn, x=column, color=color_palette[i - 1], kde=True)
    plt.xlabel(column)
    plt.title(f'{column} Histogram')

# Kolom untuk pie plot
pie_col = ['Tingkat_Pendidikan', 'Status_Pernikahan', 'Kategori_Kartu']

# Membuat subplot untuk pie plot
plt.figure(figsize=(24, 8))
for i, column in enumerate(pie_col, 1):
    plt.subplot(1, 3, i)
    dn[column].value_counts().plot.pie(colors=color_palette, autopct='%1.1f%%', startangle=140)
    plt.ylabel('')
    plt.title(f'{column}')

# Membuat subplot untuk kdeplot
plt.figure(figsize=(12, 8))
sns.kdeplot(data=dn, x='Rasio_Penggunaan_Rata-rata', fill=True)
plt.xlabel('Rasio_Penggunaan_Rata-rata')
plt.ylabel('Density')
plt.title('Rasio Penggunaan Rata-rata')

plt.tight_layout()
plt.show()
```
